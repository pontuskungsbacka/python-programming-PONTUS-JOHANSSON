###While statements 
Checking the time

```python
from datetime import datetime # Importing the datetime module to work with date and time
```

```python
datetime.now().second # Getting the current seconds
```

```python
datetime.now().second # Getting the current second + 2 seconds extra
```

##### While loops

```python
wait_until = (datetime.now().second + 2) % 60  # Calculate the target second, wrapping around at 60 
while datetime.now().second != wait_until:
    print("Still waiting...")
print(f"It's now {wait_until} seconds")  # Print the target second
```

##### Pass

```python
wait_until = (datetime.now().second + 2) % 60  # Calculate the target second, wrapping around at 60 
while datetime.now().second != wait_until:
    pass  # Busy-wait until the current second matches the target second
#Pass help us to avoid errors when we have an empty loop, and not wanting something to happen
print(f"It's now {wait_until} seconds")  # Print the target second
```

##### Break

```python
wait_until = (datetime.now().second + 2) % 60  # Calculate the target second, wrapping around at 60 
while True:
    if datetime.now().second == wait_until: # instead of while, use if. Otherwise it will be an infinite loop
        print(f"It's now {wait_until} seconds")  # Print the target second
        break # Break out of the loop when the condition is met
```

##### Continue

```python
wait_until = (datetime.now().second + 2) % 60  # Calculate the target second, wrapping around at 60 
while True:
    if datetime.now().second < wait_until: # instead of while, use if. Otherwise it will be an infinite loop
        continue # Skip the rest of the loop and start the next iteration
    break # Break out of the loop when the condition is met
print(f"It's now {wait_until} seconds")  # Print the target second
```
